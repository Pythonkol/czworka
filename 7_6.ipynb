```python
import numpy as np

initial_price = 120_000
growth_rate = 0.05
years = 5

final_price = initial_price * (1 + growth_rate) ** years
print(f"Cena mieszkania za 5 lat: {final_price:.2f} zł")
```

```python

!pip install numpy_financial
import numpy_financial as npfinancial

rate = 0.12 / 12
nper = years * 12

monthly_payment = -npfinancial.pmt(rate, nper, 0, final_price)
print(f"Wpłata miesięczna: {monthly_payment:.2f} zł")

import matplotlib.pyplot as plt

# Ceny mieszkań w czasie (liniowy wzrost po miesiącu)
months = np.arange(nper)
price_trend = initial_price*(1 + growth_rate)**(months / 12)

# Lokata zmiana:
investment_value = npfinancial.fv(rate, months, -monthly_payment, 0)

plt.plot(months, price_trend, label='Cena mieszkania')
plt.plot(months, investment_value, label='Lokata')
plt.xlabel("Miesiąc")
plt.ylabel("Wartość (zł)")
plt.legend()
plt.title("Cena mieszkania wobec lokaty")
plt.grid(True)
plt.tight_layout()
plt.show()
```
